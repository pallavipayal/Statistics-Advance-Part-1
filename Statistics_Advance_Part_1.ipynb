###1.What is a random variable in probability theory?

###2.What are the types of random variables?

###3.What is the difference between discrete and continuous distributions?

=> The key difference between discrete and continuous distributions lies in the nature of the values that the random variables can take:

Discrete Distributions:

Values:
These distributions deal with random variables that can only take on a countable number of distinct values. This means you can list the possible values, even if the list is infinitely long.
Think of "counting" things.

 Continuous Distributions:

Values:

Think of "measuring" things.

Probability:
Probabilities are assigned to ranges of values, not to individual values. The probability of a continuous random variable taking on a precise, specific value is essentially zero.
They are described by probability density functions (PDFs).



###4.What are probability distribution functions (PDF)?

=> A probability distribution describes the likelihood of different outcomes in a random experiment.
 It essentially tells you how probabilities are distributed across the possible values of a random variable

or Continuous Random Variables:
The term "probability density function" (PDF) is used.
A PDF, often denoted as f(x), represents the relative likelihood of a continuous random variable taking on a given value.
Crucially, the PDF itself doesn't give you the probability of a specific value; instead, the probability of a variable falling within a certain range is determined by calculating the area under the PDF curve over that range.
In other words, probabilities are found by integrating the PDF.

Cumulative Distribution Function (CDF): Â  The CDF gives the probability that a random variable will take on a value less than or equal to a specific value.

It ranges from 0 to 1.
It can be used for both discrete and continuous random variables

###13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

=> The Central Limit Theorem (CLT) is a cornerstone of probability theory and statistics. It states that, under certain conditions, the distribution of the sample mean (or sum) of a large number of independent, identically distributed random variables will be approximately normally distributed, regardless of the original distribution's shape.

###14.  How does the Central Limit Theorem relate to the normal distribution?

=> Sample Mean Distribution:
The CLT describes the shape of the sampling distribution of the mean. This is the distribution you'd get if you took many samples from a population, calculated the mean of each sample, and then plotted those means. The CLT says this distribution will be approximately normal.
Standard Error:
The CLT also specifies the standard deviation of this sampling distribution, which is called the standard error. It's calculated as the population standard deviation divided by the square root of the sample size. The normal distribution of the sample means is centered on the populations true mean, and has a standard deviation of the standard error.
Statistical Inference:
Because of the CLT, we can use the normal distribution to calculate probabilities and construct confidence intervals for population means, even if the population itself isn't normally distributed. This is fundamental to statistical inference.

Large Sample Sizes:

This is because, according to the Central Limit Theorem, the distribution of sample means approaches a normal distribution as the sample size increases.
Known Population Standard Deviation:

Hypothesis Testing Steps:
In hypothesis testing, the Z-statistic is used to calculate a test statistic, which is then compared to a critical value from the standard normal distribution.
This comparison helps determine whether to reject the null hypothesis (the hypothesis of no difference) in favor of the alternative hypothesis.
Determining Statistical Significance:
By calculating the Z-statistic and comparing it to critical values, researchers can determine the statistical significance of their findings.
This helps them assess the likelihood that the observed results are due to chance.
Z-tests for proportions:
Z tests are also used to compare proportions of different groups.

###17.What are point estimates and interval estimates in statistics?

###18.What is the significance of confidence intervals in statistical analysis?

=> ignificance:

Quantifying Uncertainty:
Confidence intervals address the inherent uncertainty in estimating population parameters from sample data.
They acknowledge that sample statistics are subject to random variation and may not perfectly reflect the true population values.
Providing a Range of Plausible Values:
Instead of offering a single point estimate, confidence intervals provide a range within which the population parameter is likely to lie.
This range gives a more realistic picture of the parameter's possible values.
Expressing Confidence Level:
Confidence intervals are associated with a confidence level (e.g., 95%, 99%).
This level indicates the percentage of times that the interval is expected to contain the true population parameter if the sampling process were repeated many times.
Facilitating Informed Decision-Making:
Confidence intervals aid in making informed decisions by providing a sense of the precision and reliability of estimates.
They help assess the practical significance of findings and avoid over-interpreting point estimates.
Hypothesis Testing and Statistical Inference:
Confidence intervals are closely related to hypothesis testing.
If a hypothesized value falls outside the confidence interval, it suggests evidence against the null hypothesis.
Confidence intervals are an important tool in statistical inference.
Assessing Statistical Significance:
While p-values are often used for statistical significance, confidence intervals provide additional context. For example, a confidence interval that is very wide indicates that although a result might be statistically significant, the actual effect could be very small.
Practical Significance:
Confidence intervals also help to show practical significance. A result might be statistically significant, but the confidence interval allows a researcher to see if the range of plausible values are actually meaningful in the real world.

###21.What are the assumptions for applying the Central Limit Theorem?

=> Assumptions:

Independence:
The observations in the sample must be independent of each other. This means that the value of one observation should not influence the value of another.
In practical terms, this often implies that the data is collected through random sampling.
Identical Distribution:
The observations in the sample must be drawn from the same probability distribution.
This means that all data points should have the same underlying characteristics.
Finite Variance:
The population from which the samples are drawn must have a finite variance.
In simpler terms, the data should not be infinitely spread out.
Sample Size (n):
While not a strict assumption, the CLT works best with sufficiently large sample sizes.
"Large" is somewhat subjective, but a common rule of thumb is that n â¥ 30 is generally considered large enough for the CLT to hold. However, for distributions that are already somewhat symmetrical, the CLT can work with smaller sample sizes. For strongly skewed distributions, larger sample sizes may be needed.

```python
#3.
def bernoulli_pdf(x, p):
    """
    Calculate the Probability Distribution Function (PDF) of a Bernoulli distribution.

    Parameters:
    - x: The outcome (0 or 1).
    - p: The probability of success (0 <= p <= 1).

    Returns:
    - The probability P(X = x).
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if p < 0 or p > 1:
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)


p = 0.7
x = 1

print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")

x = 0
print(f"P(X = {x}) = {bernoulli_pdf(x, p)}")
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a given dataset and plot the original data alongside its Z-scores.

    Parameters:
    - data: A list or NumPy array of numerical values.
    """
    data = np.asarray(data)

    z_scores = zscore(data)

    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.plot(data, marker='o', linestyle='-', color='b', label='Original Data')
    plt.xlabel('Index')
    plt.ylabel('Value')
    plt.title('Original Data')
    plt.grid(True)
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(z_scores, marker='o', linestyle='-', color='r', label='Z-Scores')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.title('Z-Scores')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return z_scores

data = [10, 20, 20, 30, 40, 50, 60, 70, 80, 90]
z_scores = calculate_and_plot_z_scores(data)
print("Z-Scores:", z_scores)
```

```python
#13
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def binomial_distribution(n, p, size):
    random_vars = np.random.binomial(n, p, size)

    prob = binom.pmf(random_vars, n, p)

    print("Random Variables: ", random_vars)
    print("Corresponding Probabilities: ", prob)

    plt.figure(figsize=(8, 6))
    plt.hist(random_vars, bins=np.arange(-0.5, n+1.5, 1), density=True, alpha=0.7, color='skyblue', edgecolor='black', label="Random Variables")

    x = np.arange(0, n+1)
    y = binom.pmf(x, n, p)
    plt.plot(x, y, 'ro-', label="Binomial PMF")

    plt.title(f"Binomial Distribution PMF (n={n}, p={p})")
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid(True)
    plt.show()

n = 10
p = 0.5
size = 1000

binomial_distribution(n, p, size)
```

```python
#14
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(X, mean, std_dev):
    """
    Calculate the Z-score for a given data point X.
    """
    return (X - mean) / std_dev

def plot_comparison(X, mean, std_dev):
    """
    Plot the standard normal distribution and show the Z-score comparison.
    """
    Z = calculate_z_score(X, mean, std_dev)

    x_values = np.linspace(-4, 4, 1000)
    y_values = norm.pdf(x_values, loc=0, scale=1)

    plt.plot(x_values, y_values, label="Standard Normal Distribution", color='blue')

    plt.axvline(Z, color='red', linestyle='--', label=f'Z-score = {Z:.2f}')
    plt.fill_between(x_values, y_values, where=(x_values <= Z), color='red', alpha=0.2)

    plt.title(f"Standard Normal Distribution with Z-score for X={X}")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    return Z

X = 75
mean = 70
std_dev = 5

Z = calculate_z_score(X, mean, std_dev)
print(f"Z-score for X = {X}: {Z:.2f}")

plot_comparison(X, mean, std_dev)
```

```python
#15
import numpy as np
import scipy.stats as stats

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a sample data against a population mean and standard deviation.

    Parameters:
    sample_data (array-like): The sample data.
    population_mean (float): The population mean (under the null hypothesis).
    population_std (float): The population standard deviation.
    alpha (float): The significance level (default is 0.05).

    Returns:
    result (str): The conclusion of the hypothesis test ("Reject H0" or "Fail to reject H0").
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    z_critical = stats.norm.ppf(1 - alpha / 2)

    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    if abs(z_stat) > z_critical:
        result = "Reject H0"
    else:
        result = "Fail to reject H0"

    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_stat}")
    print(f"Critical Z-value: {z_critical}")
    print(f"P-value: {p_value}")
    print(f"Result: {result}")

    return result

sample_data = np.array([85, 87, 90, 92, 88, 85, 93, 84, 89, 91])

population_mean = 90
population_std = 5

z_test(sample_data, population_mean, population_std)
```

```python
#16
import numpy as np
import scipy.stats as stats

def confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for the sample data.

    Parameters:
    sample_data (array-like): The sample data.
    confidence_level (float): The confidence level (default is 0.95).

    Returns:
    (tuple): The confidence interval (lower bound, upper bound).
    """
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    sample_size = len(sample_data)

    df = sample_size - 1

    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

    se = sample_std / np.sqrt(sample_size)

    margin_of_error = t_critical * se

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

sample_data = np.array([15, 17, 19, 20, 22, 24, 23, 18, 21, 20])

lower, upper = confidence_interval(sample_data, confidence_level=0.95)
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#17
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev, x_range=(-5, 5), num_points=1000):
    """
    Calculate and plot the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    - mean: The mean of the normal distribution.
    - std_dev: The standard deviation of the normal distribution.
    - x_range: The range of x-values to plot the PDF over.
    - num_points: The number of points to use for the x-axis (default is 1000).
    """
    x_values = np.linspace(x_range[0], x_range[1], num_points)

    pdf_values = norm.pdf(x_values, loc=mean, scale=std_dev)

    plt.figure(figsize=(8, 6))
    plt.plot(x_values, pdf_values, label=f'Normal Distribution\nMean={mean}, Std Dev={std_dev}', color='blue')
    plt.fill_between(x_values, pdf_values, color='skyblue', alpha=0.4)

    plt.title(f"Probability Density Function (PDF) of Normal Distribution\nMean = {mean}, Std Dev = {std_dev}")
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

mean = 0
std_dev = 1
plot_normal_pdf(mean, std_dev)
```

```python
#18
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lmbda, max_k):
    """
    Plot the CDF of a Poisson distribution.

    Parameters:
    - lmbda: The average rate of occurrences (lambda).
    - max_k: The maximum value of k (the number of occurrences) to plot.
    """
    k_values = np.arange(0, max_k + 1)

    cdf_values = poisson.cdf(k_values, lmbda)

    plt.figure(figsize=(8, 6))
    plt.step(k_values, cdf_values, where='post', label=f'Poisson CDF (Î»={lmbda})', color='blue')
    plt.fill_between(k_values, cdf_values, color='skyblue', alpha=0.4)
    plt.title(f"Cumulative Distribution Function (CDF) of Poisson Distribution\nÎ» = {lmbda}")
    plt.xlabel('Number of occurrences (k)')
    plt.ylabel('Cumulative Probability')
    plt.xticks(k_values)
    plt.legend()
    plt.grid(True)
    plt.show()

def poisson_cdf_value(lmbda, k):
    """
    Calculate and print the CDF value for a specific k.

    Parameters:
    - lmbda: The average rate of occurrences (lambda).
    - k: The number of occurrences.
    """
    cdf_val = poisson.cdf(k, lmbda)
    print(f"CDF for k = {k} (Î» = {lmbda}): {cdf_val:.4f}")
    return cdf_val

lmbda = 3
max_k = 10

plot_poisson_cdf(lmbda, max_k)

k = 5
poisson_cdf_value(lmbda, k)
```

```python
#19
import numpy as np
import matplotlib.pyplot as plt

def simulate_uniform_distribution(a, b, size=10000):
    """
    Simulate a random variable using a continuous uniform distribution.

    Parameters:
    - a: The lower bound of the distribution.
    - b: The upper bound of the distribution.
    - size: The number of random variables to simulate.

    Returns:
    - A numpy array of simulated random variables.
    """
    return np.random.uniform(a, b, size)

def calculate_expected_value(a, b):
    """
    Calculate the theoretical expected value of a uniform distribution.

    Parameters:
    - a: The lower bound of the distribution.
    - b: The upper bound of the distribution.

    Returns:
    - The expected value (mean) of the distribution.
    """
    return (a + b) / 2


a = 2
b = 8
random_variables = simulate_uniform_distribution(a, b, size=10000)

theoretical_expected_value = calculate_expected_value(a, b)

sample_expected_value = np.mean(random_variables)

print(f"Theoretical Expected Value: {theoretical_expected_value}")
print(f"Sample Expected Value (from simulation): {sample_expected_value}")

plt.figure(figsize=(8, 6))
plt.hist(random_variables, bins=50, density=True, alpha=0.6, color='b')
plt.axvline(theoretical_expected_value, color='r', linestyle='dashed', linewidth=2, label=f'Theoretical Mean = {theoretical_expected_value}')
plt.title("Simulated Uniform Distribution")
plt.xlabel("Random Variable Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#20
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    - data1: First dataset (array-like).
    - data2: Second dataset (array-like).
    """
    std_dev_1 = np.std(data1)
    std_dev_2 = np.std(data2)

    print(f"Standard Deviation of Dataset 1: {std_dev_1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev_2:.2f}")

    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.hist(data1, bins=20, alpha=0.7, color='blue', edgecolor='black')
    plt.title(f"Histogram of Dataset 1 (Std Dev = {std_dev_1:.2f})")
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.subplot(1, 2, 2)
    plt.hist(data2, bins=20, alpha=0.7, color='green', edgecolor='black')
    plt.title(f"Histogram of Dataset 2 (Std Dev = {std_dev_2:.2f})")
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(8, 6))
    plt.boxplot([data1, data2], vert=True, patch_artist=True, labels=['Dataset 1', 'Dataset 2'])
    plt.title("Boxplot Comparison of Datasets")
    plt.ylabel('Values')
    plt.show()

dataset_1 = np.random.normal(loc=0, scale=2, size=1000)
dataset_2 = np.random.normal(loc=0, scale=5, size=1000)
compare_standard_deviations(dataset_1, dataset_2)
```

```python
#21
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a given dataset.

    Parameters:
    - data: The dataset (array-like).

    Returns:
    - range_value: The range of the dataset.
    - iqr_value: The interquartile range (IQR) of the dataset.
    """
    range_value = np.max(data) - np.min(data)

    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)

    iqr_value = Q3 - Q1

    return range_value, iqr_value

np.random.seed(0)
mean = 0
std_dev = 1
size = 1000
data = np.random.normal(loc=mean, scale=std_dev, size=size)

range_value, iqr_value = calculate_range_and_iqr(data)

print(f"Range of the dataset: {range_value:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr_value:.2f}")
```

```python
#22
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization (standardization) to a dataset.

    Parameters:
    - data: The dataset (array-like).

    Returns:
    - normalized_data: The Z-score normalized dataset.
    """
    mean = np.mean(data)
    std_dev = np.std(data)

    normalized_data = (data - mean) / std_dev

    return normalized_data

def plot_comparison(original_data, normalized_data):
    """
    Plot the original and normalized datasets for comparison.

    Parameters:
    - original_data: The original dataset.
    - normalized_data: The Z-score normalized dataset.
    """
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=20, alpha=0.7, color='blue', edgecolor='black')
    plt.title("Original Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=20, alpha=0.7, color='green', edgecolor='black')
    plt.title("Z-score Normalized Data")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()

np.random.seed(0)
mean = 50
std_dev = 15
size = 1000

original_data = np.random.normal(loc=mean, scale=std_dev, size=size)

normalized_data = z_score_normalization(original_data)

plot_comparison(original_data, normalized_data)

print(f"Original Data - Mean: {np.mean(original_data):.2f}, Std Dev: {np.std(original_data):.2f}")
print(f"Normalized Data - Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")
```

```python
#23
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a given dataset.

    Parameters:
    - data: The dataset (array-like).

    Returns:
    - skewness: The skewness of the dataset.
    - kurtosis: The kurtosis of the dataset.
    """
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    return skewness, kurtosis

np.random.seed(0)
mean = 0
std_dev = 1
size = 1000

data = np.random.normal(loc=mean, scale=std_dev, size=size)

skewness, kurtosis = calculate_skewness_and_kurtosis(data)

print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis: {kurtosis:.4f}")

plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')
plt.title(f"Normal Distribution\nSkewness: {skewness:.4f}, Kurtosis: {kurtosis:.4f}")
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
